```python
r = "CAGTGGTGTCCGAGAAG"	
q= "GGGCGGTACT---GAGC"
```

```python
def check_wobble(r, q, wob=2):
    """   
    Check wobble pairing between r and q
    r: reference sequence
    q: query sequence
    wob: maximum wobble pairing allowed
    """ 
    print(r, q)
    print(r.count('-'), q.count('-'), len(r)!=len(q))
    if r.count('-') or q.count('-') or len(r)!=len(q): # cannot avoid gap to make gapmer
        return False, None
    #mismatch 위치학인
    diffidx = [i for i in range(len(r)) if r[i]!=q[i]]
    print(diffidx)
    if len(diffidx) > wob:
        return False, None
    # mismatch 위치에 따라 분류 
    cl1,cl2,cl3,cl4=[],[],[],[]
    for idx in diffidx:
        if (r[idx]=='C' and q[idx]=='T') or (r[idx]=='A' and q[idx]=='G'): # GU pairing human
            cl1.append(idx)
        if (r[idx]=='T' and q[idx]=='C') or (r[idx]=='G' and q[idx]=='A'): # GU pairing other
            cl2.append(idx)
        if (r[idx]=='C' and q[idx]!='G'): # I nonG pairing human
            cl3.append(idx)
        if (r[idx] not in 'GC' and q[idx]!='G'): # I nonG pairing other
            cl4.append(idx)

    if set(cl1+cl2+cl3+cl4) != set(diffidx):
        return False, None

    return True, {'GU_humanC':cl1,'GU_otherC':cl2,'I_humanC':cl3,'I_otherwise':cl4}
```

```python
check_wobble(r, q, wob=2)
```
