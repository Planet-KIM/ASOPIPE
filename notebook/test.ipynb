```python
import sys
sys.path.append("../")  # 상위 디렉토리 추가
from asopipe.main import ASOdesign
import multiprocessing as mp
import time
# ────────────────────────────
# ──────────────────────────────────────────────────────────────
# 3) 실행 예시  ────────────────────────────────────────────────
if __name__ == "__main__":
    mp.set_start_method("spawn", force=True)   # macOS/Linux 안전

    t0   = time.time()
    aso  = ASOdesign(transid="NM_002415",
                     query_assembly=["mm39","mm10"],
                     ref_assembly="hg38",
                     tile_length=17)

    result = aso.process_main(chunk_division=5, max_workers=4, wobble=2, to_df=True)
    print("Elapsed:", time.time() - t0, "sec")
```

```python
from asopipe.utils.align.maf_th import check_wobble
wobble_pair = []
wobble=0
for idx,(human, other, _locus) in enumerate(zip(result["mm39"]["hg38"], result["mm39"]["mm39"], result["mm39"]["Locus"])):
    #print(idx)
    if '-' in human or '-' in other or len(human) != len(other):
        #print(idx,False, None)
        pass
    if _locus =="chr22:23894505-23894521-":
        print(idx,human, other, _locus, [i for i, (rb, qb) in enumerate(zip(human, other)) if rb != qb])
    if idx == 99:
        print(human, other, _locus, [i for i, (rb, qb) in enumerate(zip(human, other)) if rb != qb])
    # 2) 불일치 위치 파악
    #print([i for i, (rb, qb) in enumerate(zip(human, other)) if rb != qb])
    wobble_pair.append(check_wobble(r=human, q=other, wob=wobble, anti_strand="-"))
```

```python
import pandas as pd

df = pd.DataFrame(wobble_pair)
for i, value in enumerate(wobble_pair):
    x, y = value
    if y != None:
        print(f"Index {i}: {y}")
```
